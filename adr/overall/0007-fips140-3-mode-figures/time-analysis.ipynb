```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("data/sign/fips-mode-off-ecdsa.txt") as f:
    lines = f.readlines()
sign_fips_mode_off = [float(line.strip().split(" ")[-1][:-3]) for line in lines]
with open("data/sign/fips-mode-on-ecdsa.txt") as f:
    lines = f.readlines()
sign_fips_mode_on = [float(line.strip().split(" ")[-1][:-3]) for line in lines]
```

```python
plt.scatter(np.arange(len(sign_fips_mode_off)), sign_fips_mode_off, label="FIPS 140-3 mode off",
            marker="x", color="blue")
plt.scatter(np.arange(len(sign_fips_mode_on)), sign_fips_mode_on, label="FIPS 140-3 mode on",
            marker="x", color="red")
plt.xlabel("Test cases")
plt.ylabel("Time in microseconds")
plt.title("Time Analysis of FIPS 140-3 Mode for SignCertificate() with ECDSA P-521")
plt.legend()
plt.ylim(bottom=0)
plt.tight_layout()
plt.show()
```
